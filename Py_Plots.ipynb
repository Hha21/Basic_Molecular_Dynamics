```python
import pandas as pd
import matplotlib.pyplot as plt

filename = "particles.txt"
df = pd.read_csv(filename, delim_whitespace=True)

particle_0 = df[df["ID"] == 0]
particle_1 = df[df["ID"] == 1]

plt.figure(figsize=(8, 6))
plt.plot(particle_0["X"], particle_0["Y"], label="Particle 1", linestyle="-", color="r")
plt.plot(particle_1["X"], particle_1["Y"], label="Particle 2",  linestyle="-", color="b")

plt.xlim([0, 20])
plt.ylim([0,20])

plt.xlabel("X Position (Angstroms)")
plt.ylabel("Y Position (Angstroms)")
plt.title("Particle Motion - TEST CASE 6")
plt.legend()
plt.grid()

plt.show()
```

```python
file_path = "kinetic_energy.txt"  
data = pd.read_csv(file_path, delim_whitespace=True)

plt.figure(figsize=(10, 5))
plt.plot(data["TIME"], data["KE"], label="Kinetic Energy", color="r")

#plt.xlim([0,50])
# plt.ylim([0,6])
plt.xlabel("Time")
plt.ylabel("Kinetic Energy")
plt.title("Kinetic Energy vs. Time")
plt.legend()
plt.grid()

plt.show()
```
